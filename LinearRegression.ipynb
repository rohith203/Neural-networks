```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
class LinearRegression:
    def init_weights(self, s):
        np.random.seed(11)
        self.W = np.random.randn(s[1],1)
        self.cost_arr = []
        self.cost = 0
        self.gradient = 0
        
    def get_cost(self, X, y):
        total_cost = 0
        for i in range(X.shape[0]):
            total_cost += np.square(self.get_h_i(X, i)-y[i])
        return (0.5)*total_cost
    
    def add_bias(self, X):
        bias = np.ones((X.shape[0],1))
        return np.concatenate((bias,X), axis=1)
    
    def get_h_i(self, X, i):
        h_i = 0
        for j in range(X.shape[1]):
            h_i += X[i][j]*self.W[j][0]
        return h_i
    
    def batch_gradient_descent(self, X, y, alpha):
        W_new = self.W.copy()
        for j in range(X.shape[1]):
            temp = 0
            for i in range(X.shape[0]):
                temp += (self.get_h_i(X, i) - y[i])*X[i][j]
            W_new[j][0] = self.W[j][0] - alpha*temp
        return W_new
        
    def train(self, X, y, alpha, num_iter=100):
        X = self.add_bias(X)
        self.init_weights(X.shape)
        for iteration in range(num_iter):
            self.W = self.batch_gradient_descent(X,y,alpha)
            self.cost_arr.append(self.get_cost(X, y))
        return self.cost_arr
        
    def test(self,X):
        X = self.add_bias(X)
        y_pred = np.ones(X.shape[0])
        for i in range(X.shape[0]):
            for j in range(X.shape[1]):
                y_pred[i] += X[i][j]*self.W[j][0]
        return y_pred
```

```python
model = LinearRegression()
```

```python
X = np.linspace(0,10,500).reshape(-1,1).astype(float)
# y = 2*X.reshape(-1,) + 3*np.random.randn(X.shape[0],)
y = 2*X.reshape(-1,)
```

```python
arr = model.train(X,y,0.0001,250)
```
